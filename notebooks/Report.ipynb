## üü¶ Executive Summary

This report outlines customer purchasing behavior and the performance impact of a recent store layout trial within the Chips category. It is based on our analysis across two key dimensions:

* **Customer Segment Insights**: Identifying who buys chips and what drives their purchasing behavior.
* **Store Trial Assessment**: Evaluating the trial outcomes from three stores (77, 86, 88) against control benchmarks.

Our recommendations are grounded in data and support the next phase of strategic planning for the category. The key findings are as follows:

* The Chips category is driven by **volume-focused segments**, particularly Budget-conscious families and young consumers.
* **Store 77** demonstrated a **significant increase in performance** post-layout trial, while Stores 86 and 88 showed no meaningful uplift.
* We recommend **extending the trial layout** to stores that mirror Store 77‚Äôs customer profile and delaying broader rollout until further testing.

## üîç Task 1: Customer Behavior & Segment Analysis

### 1. Key Purchasing Segments

* **Older Families (Budget)** and **Young Singles/Couples (Mainstream)** represent the highest contribution to overall sales and product quantity.
* **Retirees (Mainstream & Premium)** exhibit moderate spend and lower volume but consistent behavior.

### 2. Purchasing Trends

* Budget-conscious customers drive high volumes through frequent purchases.
* **Singles/Couples prefer larger pack sizes (\~185g)**, likely influenced by convenience.
* Families tend to buy smaller or standard pack sizes but in greater quantities.

### 3. Insight-Driven Opportunities

| Segment                    | Preference               | Opportunity                         |
| -------------------------- | ------------------------ | ----------------------------------- |
| Older Families (Budget)    | High volume, low price   | Value packs, bundled offers         |
| Young Singles (Mainstream) | Large packs, convenience | Premium-brand upselling             |
| Retirees (Premium)         | Brand loyalty            | High-margin small packs, promotions |

## üß™ Task 2: Store Trial Performance Analysis

### 1. Methodology

* We selected **control stores** for each trial store based on **Pearson correlation** and **magnitude distance scoring** using **pre-trial sales data**.
* The trial period covered **February to April 2019**.

### 2. Trial Results

| Store    | Correlation with Control | Trial Sales Uplift | Assessment                  |
| -------- | ------------------------ | ------------------ | --------------------------- |
| Store 77 | 0.92                     | +5.2%              | ‚úÖ Statistically significant |
| Store 86 | 0.89                     | -0.3%              | ‚ùå Not significant           |
| Store 88 | 0.87                     | -1.1%              | ‚ùå Not significant           |

### 3. Interpretation

* **Store 77** showed a meaningful uplift, primarily driven by an **increase in transaction volume**, not customer count.
* **Stores 86 and 88** did not demonstrate improvements above their respective control benchmarks.
